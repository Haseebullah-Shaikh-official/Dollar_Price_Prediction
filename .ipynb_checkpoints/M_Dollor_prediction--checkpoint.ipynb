```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns 

df = pd.read_csv("10y_USD_PKR Historical_Data.csv", sep=",")
df
```

```python
df.shape
```

```python
df['Date']=pd.to_datetime(df['Date'], errors='coerce')
df['Price']=pd.to_numeric(df['Price'], errors='coerce')
```

```python
df.shape
```

```python
df.info()
```

```python
df = df[['Date','Price']]
```

```python
df = df.set_index("Date")
df = df.sort_index()
df
```

```python
df.describe()
```

```python
plt.figure(figsize=(20,10))
sns.lineplot(data=df, linewidth=1)
plt.title('Dollor Price')
plt.show()
```

```python
df.hist()
plt.show()

#The diagram defines, how frequent each value occurs 
```

```python
#Checking Stationary in data 

from statsmodels.graphics.tsaplots import plot_acf
plot_acf(df)
```

```python
price_diff = df.diff(periods=1)
price_diff
```

```python
price_diff = price_diff[1:]
```

```python
price_diff
```

```python
plot_acf(price_diff)
```

```python
plt.figure(figsize=(20,10))
sns.lineplot(data=price_diff, linewidth=1)
plt.title('Dollor Price')
plt.show()
```
